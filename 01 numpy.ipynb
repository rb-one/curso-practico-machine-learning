## Datos con semi tabulados con cabeceras

```
cabeceras =  [('nombre', 'S10'), ('edad', int)]
datos = [('Juan', 10), ('Maria', 70), ('Javier', 42), ('Samuel', 15)]

# Creamos el arreglo
usuarios = np.array(datos, dtype=cabeceras)

# Hacemos sort de los datos
np.sort(usuarios, order='edad')
```

## Rangos de numeros con Arange

```
# Esto funciona como range normal
np.arange(0,21,2)
```
